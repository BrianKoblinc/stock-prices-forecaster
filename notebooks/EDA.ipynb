# EDA

```python
# to handle datasets
import pandas as pd
import numpy as np
# for plotting
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load dataset
data = pd.read_csv("portfolio_data.csv")

# rows and columns of the data
print(data.shape)

# visualise the dataset
data.head()
```

The portfolio dataset contains 1520 rows, that is, days, and 4 columns, i.e., variables.

The 4 are predictive variables and also target variables.

## Analysis

**We will analyse the following:**

1. The target variables
2. Variables types (categorical and numerical)
3. Missing data


## Target

```python
# histogran to evaluate target distribution

fig, axs = plt.subplots(2, 2, figsize=(7,5))

data['AMZN'].hist(bins=20, density=True, ax=axs[0, 0])
axs[0, 0].set_title('AMZN')
axs[0, 0].set_ylabel('Rate of days')
axs[0, 0].set_xlabel('Stock Price')

data['DPZ'].hist(bins=20, density=True, ax=axs[0, 1])
axs[0, 1].set_title('DPZ')
axs[0, 1].set_ylabel('Rate of days')
axs[0, 1].set_xlabel('Stock Price')

data['BTC'].hist(bins=20, density=True, ax=axs[1, 0])
axs[1, 0].set_title('BTC')
axs[1, 0].set_ylabel('Rate of days')
axs[1, 0].set_xlabel('Stock Price')

data['NFLX'].hist(bins=20, density=True, ax=axs[1, 1])
axs[1, 1].set_title('NFLX')
axs[1, 1].set_ylabel('Rate of days')
axs[1, 1].set_xlabel('Stock Price')

plt.tight_layout()
plt.show()
```

```python
# Prices vs time

fig, axs = plt.subplots(2, 2, figsize=(7,5))

axs[0, 0].plot(data['Date'], data['AMZN'])
axs[0, 0].set_title('AMZN')
axs[0, 0].set_ylabel('Stock Price')
axs[0, 0].set_xlabel('Date')
axs[0, 0].set_xticks(np.arange(0, len(data['Date']), 180))
axs[0, 0].tick_params(labelrotation=45)

axs[0, 1].plot(data['Date'], data['DPZ'])
axs[0, 1].set_title('DPZ')
axs[0, 1].set_ylabel('Stock Price')
axs[0, 1].set_xlabel('Date')
axs[0, 1].set_xticks(np.arange(0, len(data['Date']), 180))
axs[0, 1].tick_params(labelrotation=45)

axs[1, 0].plot(data['Date'], data['BTC'])
axs[1, 0].set_title('BTC')
axs[1, 0].set_ylabel('Stock Price')
axs[1, 0].set_xlabel('Date')
axs[1, 0].set_xticks(np.arange(0, len(data['Date']), 180))
axs[1, 0].tick_params(labelrotation=45)

axs[1, 1].plot(data['Date'], data['NFLX'])
axs[1, 1].set_title('NFLX')
axs[1, 1].set_ylabel('Stock Price')
axs[1, 1].set_xlabel('Date')
axs[1, 1].set_xticks(np.arange(0, len(data['Date']), 180))
axs[1, 1].tick_params(labelrotation=45)

plt.tight_layout()
plt.show()
```
